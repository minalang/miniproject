### 레시피 별 특징 잡아내는 함수 정의

```python
#특징 잡아내는 함수
# 1. 끓이기, 볶기, 찌기, 튀기기, 굽기 각각의 방법에서 자주 등장하는 단어의 빈도수
# 2. 위에서 가장 빈도수 높은 단어
# 3. 레시피에서 가장 많이 언급된 NNG (재료가 주로 NNG로 레시피의 주 재료를 보는 것)
import re

def suffix(document):
    features={}

    recipe_dict={}
    recipe_list=document.split(' ')

    for x in recipe_list:
      if x not in recipe_dict:
        recipe_dict[x]=0

      recipe_dict[x] +=1

    VV={}

    for word in recipe_dict.keys():
      if re.search('국물', word):
        features['국물']=recipe_dict[word]
        VV['국물']=recipe_dict[word]

      if re.search('끓이__', word):
        features['끓이__']=recipe_dict[word]
        VV['끓이__']=recipe_dict[word]

      if re.search('찌개', word):
        features['찌개']=recipe_dict[word]
        VV['찌개']=recipe_dict[word]

      if re.search('볶/', word):
        features['볶/']=recipe_dict[word]
        VV['볶/']=recipe_dict[word]
      
      if re.search('찜__', word):
        features['찜__']=recipe_dict[word]
        VV['찜__']=recipe_dict[word]

      if re.search('찌__', word):
        features['찌__']=recipe_dict[word]
        VV['찌__']=recipe_dict[word]

      if re.search('튀김__', word):
        features['튀김__']=recipe_dict[word]
        VV['튀김__']=recipe_dict[word]

      if re.search('튀기__', word):
        features['튀기__']=recipe_dict[word]
        VV['튀기__']=recipe_dict[word]

      if re.search('튀김옷', word):
        features['튀김옷']=recipe_dict[word]
        VV['튀김옷']=recipe_dict[word]

      #if re.search('전__', word):
        #features['전__']=recipe_dict[word]
        #VV['전__']=recipe_dict[word]

      if re.search('부침__', word):
        features['부침__']=recipe_dict[word]
        VV['부침__']=recipe_dict[word]

      if re.search('부치__', word):
        features['부치__']=recipe_dict[word]
        VV['부치__']=recipe_dict[word]

      if re.search('달구__', word):
        features['달구__']=recipe_dict[word]
        VV['달구__']=recipe_dict[word]

      if re.search('굽__', word):
       features['굽__']=recipe_dict[word]
       VV['굽__']=recipe_dict[word]

      if re.search('반죽', word):
       features['반죽']=recipe_dict[word]
       VV['반죽']=recipe_dict[word]

      if re.search('오븐', word):
       features['오븐']=recipe_dict[word]
       VV['오븐']=recipe_dict[word]


      #features['재료갯수(NNG)']=0
      #for word in recipe_dict.keys():
        #if re.search('(/NNG|/NNP)', word):
          #features['재료갯수(NNG)']+=1

      #NNG={}
      NN={}

      for word in recipe_dict.keys():
        if re.search('케이크', word):
          features['케이크']=recipe_dict[word]
          NN['케이크']=recipe_dict[word]

        if re.search('우유', word):
          features['우유']=recipe_dict[word]
          NN['우유']=recipe_dict[word]

        #if re.search('고기', word):
          #features['고기']=recipe_dict[word]
          #NN['고기']=recipe_dict[word]

        if re.search('스테이크', word):
          features['스테이크']=recipe_dict[word]
          NN['스테이크']=recipe_dict[word]

      #for word in recipe_dict.keys():
        #if re.search('(/NNG|/NNP)', word):
          #NNG[word]=recipe_dict[word]

      #features['재료']=sorted(NNG.items(), key=lambda item: item[1],reverse=True)[0][0]

      features['most_VV']=None
      if len(VV)>0:
        features['most_VV']=sorted(VV.items(), key=lambda item: item[1],reverse=True)[0][0]

      features['most_NN']=None
      if len(NN)>0:
        features['most_NN']=sorted(NN.items(), key=lambda item: item[1],reverse=True)[0][0]

    #sorted(final[i].items(), key=lambda item: item[1], reverse=True)
    
    return features

suffix('황태__02/NNG 구이__01/NNG 황태__02/NNG 가위__01/NNG 자르__01/VV 물__01/NNG 담그/VV 짜__02/VV 다음__01/NNG 후라이팬/NNG 식용유/NNG 포도__06/NNG 씨__01/NNG 유__02/NNG 사용하__03/VV 두르/VV 익히__02/VV 뚜껑/NNG 닫__02/VV 앞뒤/NNG 익히__01/VV 다음__01/NNG 양념장/NNG 다음__01/NNG 바르__01/VV 굽__01/VV 속__01/NNG 익__01/VV 양념장/NNG 타__02/VV 고추장/NNG 들어가__01/VV 양념장/NNG 속__01/NNG 재료__01/NNG 익__01/VV 전__08/NNG 겉__01/NNG 타__01/VV 설탕/NNG 들어가__01/VV 타__01/VV 시작하__01/VV 양념장/NNG 계량하__01/VV 마르__01/VV 북어/NNG 고추장/NNG 슬__02/VV 설탕/NNG 물엿/NNG 맛술/NNG 참기름/NNG 깨소금/NNG 다지/VV 마늘/NNG 진간장__01/NNG 슬__02/VV 물__01/NNG 슬__02/VV 정도__11/NNG 넣/VV 개__02/VV 후라이팬/NNG 애벌/NNG 익히__02/VV 황태__02/NNG 나중__01/NNG 양념장/NNG 바르__01/VV 앞뒤/NNG 굽__01/VV 다음__01/NNG 마지막/NNG 잔파/NNG 썰__01/VV 뿌리/VV 끝__01/NNG 타__01/VV 굽__01/VV 처음/NNG 애벌구이하/VV 때__01/NNG 식용유/NNG 넣/VV 고구마/NNG 죽__07/NNG 아침/NNG 찬밥__01/NNG 죽__07/NNG 쑤__01/VV 오늘/NNG 고구마/NNG 하__01/VV 하__01/VV 고구마/NNG 양파/NNG 당근/NNG 끓이__01/VV 믹서/NNG 갈__02/VV 다음__01/NNG 브로콜리/NNG 다지/VV 넣/VV 밥__01/NNG 끓이__01/VV 되__01/VV 당근/NNG 들어가__01/VV 색__03/NNG 입맛/NNG 맞__01/VV 소금__01/NNG 간__01/NNG 하__01/VV 되__01/VV 생오이/NNG 무침/NNG 죽__07/NNG 끓/VV 동안__01/NNG 오이__01/NNG 하나/NNP 썰__01/VV 무쳐내/VV 생오이/NNG 무침/NNG 먹__02/VV 새콤/NNG 달콤/NNG 무쳐내/VV 아삭하/VV 먹__02/VV 하__01/VV')
```
