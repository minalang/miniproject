```python
import pandas as pd
import numpy as np

data=pd.read_csv('/content/drive/My Drive/DataCampus/project/DATA_final_0819.csv', encoding='cp949')
data=data.iloc[:, :33]
```

```python
#재료 없는 항 삭제

for x in range(0, len(data)):
  if str(data['ingredient'][x])=='nan':
    print(x)

    data=data.drop([x], axis=0)

data
```

```python
# all_recipe

#print(df['RECIPE-1'][0])
#df.fillna('None')

import re

recipe_dict={}

for x in data.index:
  recipe_list=[]

  for i in range(1,26):

    recipe_str='RECIPE-'+str(i)

    if str(data[recipe_str][x])=='nan':
      #print(df[recipe_str][x])
      continue
    
    recipe_content=str(data[recipe_str][x])
    recipe_content=re.sub('^\d[.]', '', recipe_content)
    recipe_content=re.sub('\n', '', recipe_content)

    recipe_list.append(recipe_content)
    #print(recipe_str)
  
  recipe_dict[x]=recipe_list

#print(recipe_dict)

all_recipe={}

for x,y in recipe_dict.items():

  all_recipe[x]='\n'.join(y)
  #print('\n'.join(y))

#all_recipe
pd.DataFrame({'all_recipe':all_recipe})
#pd.concat([df['TITLE'], pd.DataFrame({'all_recipe':all_recipe})], axis=0).tail(10)
#df['TITLE']

data=pd.concat([data,pd.DataFrame({'all_recipe':all_recipe})], axis=1 )
data
```
