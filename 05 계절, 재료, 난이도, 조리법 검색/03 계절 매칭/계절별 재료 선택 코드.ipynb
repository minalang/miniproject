# 계절별 재료 선택 코드

```python
#필요한 패키지 불러오기
import pandas as pd #파일불러올 떄
import numpy as np
import os #디렉토리
from scipy import stats
```

```python
data=pd.read_excel('/content/drive/Shareddrives/말해조/정민화/07계절매칭/재철달 매칭.xlsx')
```

```python
data.head(20) #데이터 확인
```

```python
season=input("원하는 계절을 선택해주세요: ") #원하는 계절 입력으로 받고
data_mask=data['month']==season #입력값과 같은 계절의 데이터만 추출 
filtered_data = data[data_mask]
print(filtered_data['ingredient']) #추출된 계절의 재료만 출력
```
